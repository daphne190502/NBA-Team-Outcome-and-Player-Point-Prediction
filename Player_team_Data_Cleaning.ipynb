```python
columns_to_sum = [
    'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
    'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS'
]

# Columns with percentage values and their corresponding attempts
pct_columns = {
    'FG_PCT': 'FGA',
    'FG3_PCT': 'FG3A',
    'FT_PCT': 'FTA'
}

# Create new columns for weighted sum and total for each percentage column
for pct_col, attempt_col in pct_columns.items():
    NBA_games[f'{pct_col}_sum'] = NBA_games[pct_col] * NBA_games[attempt_col]
    NBA_games[f'{pct_col}_total'] = NBA_games[attempt_col]

# Aggregation dictionary for summing numeric columns and handling temporary columns
aggregations = {'TEAM_CITY': 'first'}
aggregations.update({col: 'sum' for col in columns_to_sum})

# Add entries for weighted sum and total columns
aggregations.update({f'{pct_col}_sum': 'sum' for pct_col in pct_columns})
aggregations.update({f'{pct_col}_total': 'sum' for pct_col in pct_columns})

# Use 'first' for binary/categorical and season columns
aggregations.update({'TEAM_WINS': 'first', 'IS_HOME': 'first', 'SEASON': 'first','TEAM_CITY': 'first'})

# Perform aggregation
team_game_data = NBA_games.groupby(['GAME_ID', 'TEAM_ID']).agg(aggregations).reset_index()

# Calculate weighted averages for percentage columns
for pct_col in pct_columns.keys():
    team_game_data[pct_col] = round(team_game_data[f'{pct_col}_sum'] / team_game_data[f'{pct_col}_total'], 3)

# Drop temporary columns used for calculation
team_game_data.drop(columns=[f'{pct_col}_sum' for pct_col in pct_columns] + 
                     [f'{pct_col}_total' for pct_col in pct_columns], inplace=True)
```
