```python

# Display the columns to verify the column names
print("Columns in the dataset:", team_data.columns)

# Check if the necessary columns exist
required_columns = ['GAME_ID', 'TEAM_ID', 'PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB', 'TEAM_WINS', 'STL', 'BLK', 'IS_HOME', 'TO', 'PF', 'FGA', 'eFG%', 'FGM', 'GAME_DATE_EST']
for col in required_columns:
    if col not in team_data.columns:
        raise KeyError(f'Column not found: {col}')

# Create a dictionary that maps GAME_ID to the list of teams that played in that game
game_to_teams = team_data.groupby('GAME_ID')['TEAM_ID'].apply(list).to_dict()

# Define a function to get the opponent team ID'
def get_opponent_team_id(row):
    teams = game_to_teams[row['GAME_ID']]
    return teams[1] if teams[0] == row['TEAM_ID'] else teams[0]

# Apply the function to create the OPP_TEAM_ID column
team_data['OPP_TEAM_ID'] = team_data.apply(get_opponent_team_id, axis=1)

# Verify that the OPP_TEAM_ID column has been created
if 'OPP_TEAM_ID' not in team_data.columns:
    raise KeyError('OPP_TEAM_ID column not found in the DataFrame')

# Sort by team and game date
team_data['GAME_DATE_EST'] = pd.to_datetime(team_data['GAME_DATE_EST'])
team_data = team_data.sort_values(by=['TEAM_ID', 'GAME_DATE_EST'])

# Calculate rolling averages for the last 5 games, excluding the current game for TEAM_ID
team_data['avg_pts'] = team_data.groupby('TEAM_ID')['PTS'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_fg_pct'] = team_data.groupby('TEAM_ID')['FG_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_ft_pct'] = team_data.groupby('TEAM_ID')['FT_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_fg3_pct'] = team_data.groupby('TEAM_ID')['FG3_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_ast'] = team_data.groupby('TEAM_ID')['AST'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_reb'] = team_data.groupby('TEAM_ID')['REB'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_stl'] = team_data.groupby('TEAM_ID')['STL'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_to'] = team_data.groupby('TEAM_ID')['TO'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_pf'] = team_data.groupby('TEAM_ID')['PF'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_fga'] = team_data.groupby('TEAM_ID')['FGA'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_efg_pct'] = team_data.groupby('TEAM_ID')['eFG%'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_fgm'] = team_data.groupby('TEAM_ID')['FGM'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['avg_blk'] = team_data.groupby('TEAM_ID')['BLK'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())

# Calculate win/loss streaks for TEAM_ID
def calculate_win_streaks(group):
    streak = 0
    streaks = []
    for result in group.shift(1):  # Shift by 1 to exclude the current game
        if result == 1:
            streak += 1
        else:
            streak = 0
        streaks.append(streak)
    return pd.Series(streaks, index=group.index)

team_data['win_streak'] = team_data.groupby('TEAM_ID')['TEAM_WINS'].apply(calculate_win_streaks).reset_index(level=0, drop=True)

# Calculate home/away performance for TEAM_ID
team_data['home_avg_pts'] = team_data.groupby('TEAM_ID').apply(lambda group: group[group['IS_HOME'] == 1]['PTS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['away_avg_pts'] = team_data.groupby('TEAM_ID').apply(lambda group: group[group['IS_HOME'] == 0]['PTS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['home_win_pct'] = team_data.groupby('TEAM_ID').apply(lambda group: group[group['IS_HOME'] == 1]['TEAM_WINS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['away_win_pct'] = team_data.groupby('TEAM_ID').apply(lambda group: group[group['IS_HOME'] == 0]['TEAM_WINS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)

# Calculate head-to-head features
head_to_head = team_data.groupby(['TEAM_ID', 'OPP_TEAM_ID']).agg(
    head_to_head_avg_pts=('PTS', 'mean'),
    head_to_head_win_pct=('TEAM_WINS', 'mean')
).reset_index()

# Merge head-to-head stats back to the main dataframe for home and away teams
team_data = pd.merge(team_data, head_to_head, left_on=['TEAM_ID', 'OPP_TEAM_ID'], right_on=['TEAM_ID', 'OPP_TEAM_ID'], how='left')
team_data = team_data.rename(columns={
    'head_to_head_avg_pts': 'home_head_to_head_avg_pts',
    'head_to_head_win_pct': 'home_head_to_head_win_pct'
})

team_data = pd.merge(team_data, head_to_head, left_on=['OPP_TEAM_ID', 'TEAM_ID'], right_on=['TEAM_ID', 'OPP_TEAM_ID'], how='left')
team_data = team_data.rename(columns={
    'head_to_head_avg_pts': 'away_head_to_head_avg_pts',
    'head_to_head_win_pct': 'away_head_to_head_win_pct'
})

# Calculate rolling averages for the last 5 games, excluding the current game for OPP_TEAM_ID
team_data = team_data.sort_values(by=['OPP_TEAM_ID_x', 'GAME_DATE_EST'])

team_data['opp_avg_pts'] = team_data.groupby('OPP_TEAM_ID_x')['PTS'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_fg_pct'] = team_data.groupby('OPP_TEAM_ID_x')['FG_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_ft_pct'] = team_data.groupby('OPP_TEAM_ID_x')['FT_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_fg3_pct'] = team_data.groupby('OPP_TEAM_ID_x')['FG3_PCT'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_ast'] = team_data.groupby('OPP_TEAM_ID_x')['AST'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_reb'] = team_data.groupby('OPP_TEAM_ID_x')['REB'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_stl'] = team_data.groupby('OPP_TEAM_ID_x')['STL'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_to'] = team_data.groupby('OPP_TEAM_ID_x')['TO'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_pf'] = team_data.groupby('OPP_TEAM_ID_x')['PF'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_fga'] = team_data.groupby('OPP_TEAM_ID_x')['FGA'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_efg_pct'] = team_data.groupby('OPP_TEAM_ID_x')['eFG%'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_fgm'] = team_data.groupby('OPP_TEAM_ID_x')['FGM'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
team_data['opp_avg_blk'] = team_data.groupby('OPP_TEAM_ID_x')['BLK'].transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())

# Calculate win/loss streaks for OPP_TEAM_ID
team_data['opp_win_streak'] = team_data.groupby('OPP_TEAM_ID_x')['TEAM_WINS'].apply(calculate_win_streaks).reset_index(level=0, drop=True)

# Calculate home/away performance for OPP_TEAM_ID
team_data['opp_home_avg_pts'] = team_data.groupby('OPP_TEAM_ID_x').apply(lambda group: group[group['IS_HOME'] == 1]['PTS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['opp_away_avg_pts'] = team_data.groupby('OPP_TEAM_ID_x').apply(lambda group: group[group['IS_HOME'] == 0]['PTS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['opp_home_win_pct'] = team_data.groupby('OPP_TEAM_ID_x').apply(lambda group: group[group['IS_HOME'] == 1]['TEAM_WINS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)
team_data['opp_away_win_pct'] = team_data.groupby('OPP_TEAM_ID_x').apply(lambda group: group[group['IS_HOME'] == 0]['TEAM_WINS'].shift(1).rolling(window=5, min_periods=1).mean()).reset_index(level=0, drop=True)

# Fill missing values using forward fill
team_data = team_data.fillna(method='ffill')

# If there are still NaNs (for the very first match), fill them with 0
team_data = team_data.fillna(0)


# Select relevant features for the model
features = [
    'avg_pts', 'avg_fg_pct', 'avg_ft_pct', 'avg_fg3_pct', 'avg_ast', 'avg_reb', 
    'avg_stl', 'avg_to', 'avg_pf', 'avg_fga', 'avg_efg_pct', 'avg_fgm',
    'win_streak', 'home_avg_pts', 'away_avg_pts', 'home_win_pct', 'away_win_pct', 
    'home_head_to_head_avg_pts', 'home_head_to_head_win_pct', 
    'away_head_to_head_avg_pts', 'away_head_to_head_win_pct',
    'opp_avg_pts', 'opp_avg_fg_pct', 'opp_avg_ft_pct', 'opp_avg_fg3_pct', 
    'opp_avg_ast', 'opp_avg_reb', 'opp_avg_stl', 'opp_avg_to', 'opp_avg_pf', 
    'opp_avg_fga', 'opp_avg_efg_pct', 'opp_avg_fgm',
    'opp_win_streak', 'opp_home_avg_pts', 'opp_away_avg_pts', 'opp_home_win_pct', 'opp_away_win_pct'
]

target = 'TEAM_WINS'

# Split the dataset into features (X) and target (y)
X = team_data[features]
y = team_data[target]

# Display the first few rows of the feature matrix and target variable
print(X.head())
print(y.head())
```
